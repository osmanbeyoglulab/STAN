```python
import scanpy as sc
import seaborn as sns
from matplotlib import pyplot as plt


plt.rcParams['pdf.fonttype']=42
plt.rcParams.update({'font.sans-serif':'Helvetica'})
plt.rcParams.update({'font.size': 7})
```

```python
adata=sc.read("out/stf_results.h5ad")
```

```python
plt.figure(figsize=(2,2))
plt.plot([0,1], [0,1], '-', alpha=0.3, color='grey')
sns.scatterplot(data=adata.obs, x="pred_cor_stl", y='pred_cor_stan', s=0.5, hue="n_counts", linewidth=0)
plt.xlim([0.16,0.29])
plt.ylim([0.16,0.29])
plt.ylabel("Prediction Correlation (STAN)")
plt.xlabel("Prediction Correlation (baseline)")
plt.legend(title="UMI count", loc='lower right')
plt.savefig("2b_performance_scatterplot.pdf", bbox_inches="tight")
```

```python
perf_diff=adata.obs['pred_cor_stan']-adata.obs['pred_cor_stl']
sns.scatterplot(x=perf_diff, y=adata.obs['ncounts'], s=1)
```

```python
from scipy.stats import pearsonr
pearsonr(perf_diff,adata.obs['ncounts'])
```
